```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/Kalandra/Kalandra.items.csv', delimiter=';', parse_dates=['Date'])
df = df.loc[df['Confidence'] != 'Low']
df['FirstDate'] = df.groupby('Id')['Date'].transform('min')
df.columns
```

```python
day1 = df['Date'].min()
day2_change = df.loc[(df['Date'] == day1 + np.timedelta64(1, 'D')) | (df['Date'] == day1)]
day2_change = day2_change.loc[day2_change['Value'] >= 5]
day2_change['Change'] = day2_change.groupby('Id')['Value'].pct_change()
day3_change = df.loc[(df['Date'] == day1 + np.timedelta64(3, 'D')) | (df['Date'] == df['FirstDate'])]
day3_change = day3_change.loc[day3_change['Value'] >= 5]
day3_change['Change'] = day3_change.groupby('Id')['Value'].pct_change()
day7_change = df.loc[(df['Date'] == day1 + np.timedelta64(7, 'D')) | (df['Date'] == df['FirstDate'])]
day7_change = day7_change.loc[day7_change['Value'] >= 5]
day7_change['Change'] = day7_change.groupby('Id')['Value'].pct_change()
day1
```

```python
day2_change.nlargest(10, 'Change')
```
