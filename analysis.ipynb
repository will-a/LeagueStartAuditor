```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/Kalandra/Kalandra.items.csv', delimiter=';', parse_dates=['Date'])
df['Change'] = df.groupby('Id')['Value'].pct_change()
df.columns
```

```python
day1 = df['Date'].min()
day1
day2_change = df.loc[df['Date'] == day1 + np.timedelta64(1, 'D')]
day2_change = day2_change.loc[day2_change['Value'] >= 5]
```

```python
day2_change.nlargest(10, 'Change')
```

```python
day2_change.nsmallest(10, 'Change')
```
